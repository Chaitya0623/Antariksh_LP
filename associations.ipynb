```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/predict_stock_values.csv')
df
```

```python
stock_val={}
for i in range(len(df)):
  sequence_values = df['Actual_Price_List'][i].strip('[]').split()
  stock_val[df['Company_Name'][i]] = sequence_values[-1]
stock_val
```

```python
stock_val.keys()
```

```python
import random

num_rows = 100000
rows = []
for _ in range(num_rows):
    num_elements = random.randint(1, 5)
    random_elements = random.sample(stock_val.keys(), num_elements)
    rows.append(random_elements)
```

```python
rows[0]
```

```python
data = pd.DataFrame({'Stocks': rows})
data
```

```python
# Create a new DataFrame to store the expanded data
expanded_data = []

# Iterate through the original DataFrame
for idx, row in data.iterrows():
    unique_id = idx
    for element in row['Stocks']:
        expanded_data.append([unique_id, element])

# Create the expanded DataFrame
expanded_df = pd.DataFrame(expanded_data, columns=['unique_id', 'element'])
expanded_df
```

```python
onehot = expanded_df.pivot_table(index='unique_id',columns='element', aggfunc=len, fill_value=0)
onehot = onehot>0
onehot
```

```python
onehot.to_csv('onehot.csv')
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# compute frequent items using the Apriori algorithm
frequent_itemsets = apriori(onehot, min_support = 0.001, max_len = 5, use_colnames=True)

# compute all association rules for frequent_itemsets
rules = association_rules(frequent_itemsets, min_threshold=0.002)
rules.sort_values('confidence', ascending=False).reset_index(drop=True)
```

```python
# New list for which you want to find associations
new_list = ['Atos', 'Cap Gemini']  # Replace with actual elements

# Calculate the support for the new list
support_new_list = (onehot[new_list] == 1).all(axis=1).mean()

# Find rules that involve elements from the new list
relevant_rules = rules[rules['antecedents'].apply(lambda x: any(item in new_list for item in x))]
relevant_rules = relevant_rules.sort_values('confidence', ascending=False)

# Print relevant rules and their confidence
print("Relevant Rules:")
print(relevant_rules)
```
